```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, auc
import xgboost as xgb
import lightgbm as lgb

%matplotlib inline
```

```python
#loading the 'Electrical grid stability simulated data' csv file into a Pandas DataFrame
#and viewing 20 random samples

grid_stability_df = pd.read_csv('Data_for_UCI_named.csv')
grid_stability_df.sample(20)
```

```python
#checking the DataFrame for null fields and also the Dtype of each feature

grid_stability_df.info()
```

```python
#checking for common stats of each feature

grid_stability_df.describe()
```

```python
#creating a Scaler

scaler = StandardScaler()
```

```python
#checking the number of occurences of each class of the target variable

grid_stability_df.stabf.value_counts()
```

```python
grid_stability_df.isna().sum()
```

```python
X = grid_stability_df.drop(columns=['stab', 'stabf'])
X
```

```python
y = grid_stability_df['stabf']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_train_scaled
```

```python
X_test_scaled = pd.DataFrame(scaler.fit_transform(X_test), columns=X_test.columns)
X_test_scaled
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_train_encoded
```

```python
encoder.transform(['unstable'])
```

```python
encoder.transform(['stable'])
```

```python
y_test_encoded = encoder.fit_transform(y_test)
y_test_encoded
```

```python
y_train_scaled = scaler.fit_transform(np.array(y_train_encoded).reshape(-1,1))
y_train_scaled
```

```python
y_test_scaled = scaler.fit_transform(np.array(y_test_encoded).reshape(-1,1))
y_test_scaled
```

```python
len(y_train_scaled)
```

```python
len(y_test_scaled)
```

```python
rf_clf = RandomForestClassifier(random_state=1)
```

```python
rf_clf.fit(X_train_scaled, y_train)
```

```python
rf_pred = rf_clf.predict(X_test_scaled)
```

```python
rf_accuracy_score = accuracy_score(y_test, rf_pred)
rf_accuracy_score
```

```python
rf_precision_score = precision_score(y_test, rf_pred, pos_label='stable')
rf_precision_score
```

```python
rf_precision_score = precision_score(y_test, rf_pred, pos_label='unstable')
rf_precision_score
```

```python
rf_precision_score = precision_score(y_test, rf_pred, average='weighted')
rf_precision_score
```

```python
rf_recall_score = recall_score(y_test, rf_pred, pos_label='stable')
rf_recall_score
```

```python
rf_recall_score = recall_score(y_test, rf_pred, pos_label='unstable')
rf_recall_score
```

```python
rf_recall_score = recall_score(y_test, rf_pred, average='weighted')
rf_recall_score
```

```python
rf_f1 = f1_score(y_test, rf_pred, pos_label='stable')
rf_f1
```

```python
rf_f1 = f1_score(y_test, rf_pred, pos_label='unstable')
rf_f1
```

```python
rf_f1 = f1_score(y_test, rf_pred, average='weighted')
rf_f1
```
